```python

def count_lines_and_words(file_path):
    
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
            num_lines = len(lines)
            num_words = sum(len(line.split()) for line in lines)
        return num_lines, num_words
    except FileNotFoundError:
        return f"File not found: {file_path}"
    

files={"Obama's speech":"obama_speech.txt",
       "Michelle Obama's":"michelle_obama_speech.txt",
       "Melina Trump's speech":"melina_trump_speech.txt",
       "Donald's speeh":"donald_speech.txt"}


for title, path in files.items():
    lines, words = count_lines_and_words(path)
    print(f"{title}: {lines} lines, {words} words")



#most spoken language
import json
from collections import Counter
def ten_mst_spoken_lang(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            countries_data = json.load(file)
        language_counts = Counter(
            language for country in countries_data for language in country.get('languages',[])

        )
        return language_counts.most_common(10)
    except FileNotFoundError:
        return f"file not found:{file_path}"
    
file_path="countries_data.json"

most_spkn_lang =ten_mst_spoken_lang(file_path)
if isinstance(most_spkn_lang,list):
    for count,language in most_spkn_lang:
        print(f"({language},'{count}')")
else:
    print(most_spkn_lang)

#most populated countries
import json

def ten_most_populated_con(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            countries_data = json.load(file)

       
        most_populated = sorted(
            countries_data,
            key=lambda x: x['population'],
            reverse=True
        )[:10]
        
       
        result = [{'country': country['name'], 'population': country['population']} for country in most_populated]
        return result
    
    except FileNotFoundError:
        return f"File not found: {file_path}"
    except json.JSONDecodeError:
        return f"Error decoding JSON in file: {file_path}"


file_path = "countries_data.json"
most_populated_con = ten_most_populated_con(file_path)


if isinstance(most_populated_con, list):
    print("Ten most populated countries:")
    for country_info in most_populated_con:
        print(f"{country_info}")
else:
    print(most_populated_con)

```

```python
import re
def extract_emails(file_path):
    try:
        with open(file_path,'r', encoding='utf-8') as file:
            text = file.readlines()

        incoming_emails = []
        for line in text:
            if line.startswith("From:"):
                email = re.search(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', line)
                if email:
                    incoming_emails.append(email.group())
        return incoming_emails
    except FileNotFoundError:
        return f"Error: The file {file_path} was not found."
    except Exception as e:
        return str(e)
emails_file = "email_exchanges_big.txt"
incoming_emails = extract_emails(emails_file)
print("Incoming emails:")
for email in incoming_emails:
    print(email)
   

emails_file = "email_exchanges_big.txt"
incoming_emails = extract_emails(emails_file)
print("Incoming emails:")
for email in incoming_emails:
    print(email)


```

```python
#FIND MOST COMMON WORDS

from collections import Counter
import re

def find_most_common_words(input_data, n):
    try:
        
        if "\n" not in input_data and input_data.endswith('.txt'):
            with open(input_data, 'r', encoding='utf-8') as file:
                text = file.read()
        else:
            text = input_data
        
        words = re.findall(r'\b\w+\b', text.lower())

        word_counts = Counter(words)
        
        most_common = word_counts.most_common(n)
        
        return [(freq, word) for word, freq in most_common]
    
    except FileNotFoundError:
        print(f"Error: File '{input_data}' not found.")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []


sample_text = """To be or not to be, that is the question:
Whether 'tis nobler in the mind to suffer
The slings and arrows of outrageous fortune,
Or to take arms against a sea of troubles
And by opposing end them."""
print(find_most_common_words(sample_text, 10))


```

```python

#Finding most frequent words in speeche'sample_text
import re
from collections import Counter
def find_most_frequent_words(file_path,n):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            # Tokenize and clean the text
            words = re.findall(r'\b\w+\b', text.lower())
            word_counts = Counter(words)
            most_frequent = word_counts.most_common(n)
            return [(freq, word) for word, freq in most_frequent]
    except FileNotFoundError:
        return f"File not found: {file_path}"
    except Exception as e:
        return f"An error occurred: {str(e)}"
    
#speeches file paths
Obama_speech = "obama_speech.txt"
michelle_speech = "michelle_obama_speech.txt"
trump_speech = "donald_speech.txt"
mellina_speech = "melina_trump_speech.txt"

print("Obama's Speech:")
obama_most_frequent = find_most_frequent_words(Obama_speech, 10)
if isinstance(obama_most_frequent, str):
    print(obama_most_frequent)  # Handle error messages
else:
    for freq, word in obama_most_frequent:
        print(f"{word}: {freq}")

print("\nMichelle's Speech:")
michelle_most_frequent = find_most_frequent_words(michelle_speech, 10)
if isinstance(michelle_most_frequent, str):
    print(michelle_most_frequent)
else:
    for freq, word in michelle_most_frequent:
        print(f"{word}: {freq}")

print("\nTrump's Speech:")
trump_most_frequent = find_most_frequent_words(trump_speech, 10)
if isinstance(trump_most_frequent, str):
    print(trump_most_frequent)
else:
    for freq, word in trump_most_frequent:
        print(f"{word}: {freq}")

print("\nMelina's Speech:")
melina_most_frequent = find_most_frequent_words(mellina_speech, 10)
if isinstance(melina_most_frequent, str):
    print(melina_most_frequent)
else:
    for freq, word in melina_most_frequent:
        print(f"{word}: {freq}")

```

```python
import re
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to clean text
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower().strip()

def remove_support_words(text, stop_words):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# Check text similarity
def check_text_similarity(text1, text2):
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([text1, text2])
    similarity_score = cosine_similarity(vectors[0], vectors[1])
    return similarity_score[0][0]

# Read stop words from a file
def load_stop_words(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            stop_words = file.read().splitlines()
        return set(stop_words)
    except FileNotFoundError:
        return set()

# Function to process files or strings and calculate similarity
def process_and_compare(file1, file2, stop_words_file):
    stop_words = load_stop_words(stop_words_file)

    try:
        # Read the first file
        with open(file1, 'r', encoding='utf-8') as f1:
            text1 = f1.read()
        
        # Read the second file
        with open(file2, 'r', encoding='utf-8') as f2:
            text2 = f2.read()
        
        # Clean and process the texts
        text1_clean = clean_text(text1)
        text2_clean = clean_text(text2)
        
        text1_filtered = remove_support_words(text1_clean, stop_words)
        text2_filtered = remove_support_words(text2_clean, stop_words)

        # Calculate similarity
        similarity_score = check_text_similarity(text1_filtered, text2_filtered)
        return similarity_score
    except FileNotFoundError as e:
        return f"File not found: {e.filename}"

# Paths to the speeches and stop words file
text1 = "michelle_obama_speech.txt"
text2 = "melina_trump_speech.txt"
stop_words = "stop_words.py"

# Check similarity between the speeches
similarity = process_and_compare(text1, text2, stop_words)
print(f"Similarity Score between Michelle's and Melina's speech: {similarity:.2f}")
```

```python
from collections import Counter
import re

def find_most_repeated_words(file_p,n):
    try:
        with open(file_p,'r',encoding='utf-8') as file:
            text = file.read()
            text = re.sub(r'[^\w\s]','',text.lower())

            words = text.split()
            word_counts = Counter(words)
            most_common_words = word_counts.most_common(n)

            return most_common_words
    except FileNotFoundError:
        return f"File not found:{file_p}"
    
file_p = "romeo_and_juliet.txt"
find_most_repeated_words = find_most_repeated_words(file_p, 10)

print("The most repeated words in Romeo and Juliet are:")
for word,count in find_most_repeated_words:
    print(f"{word}: {count}")
```

```python
import re
import requests

def analyze_hacker_news(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.text.splitlines()
        
        python_count = 0
        Javascript_count = 0
        java_count = 0
        
        for line in data:
            if re.search(r'\bpython\b' ,line,re.IGNORECASE):
                python_count +=1

            if re.search(r'bjavascript\b',line,re.IGNORECASE):
                Javascript_count +=1

            if re.search(r'bjava\b',line,re.IGNORECASE) and not re.search(r'bjavascript\b',line,re.IGNORECASE):
                java_count +=1

        return python_count,Javascript_count,java_count
    except Exception as e:
        return str(e)
    

url = "https://raw.githubusercontent.com/Asabeneh/30-Days-Of-Python/refs/heads/master/data/hacker_news.csv"

python_count,Javascript_count,java_count=analyze_hacker_news(url)
print(f"lines containing 'python' or 'Python' : {python_count}")
print(f"Lines containing 'JavaScript','javascript' or 'Javascript':{Javascript_count}")
print(f"Lines containing 'java' and not 'javascript': {java_count}")


```
