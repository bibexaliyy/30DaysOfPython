- Use for loop to iterate from 0 to 100 and print the sum of all evens and the sum of all odds.The sum of all evens is 2550. And the sum of all odds is 2500.

```python
total_sum = 0
for i in range(101):
    total_sum += i
print(f"The sum of all numbers is {total_sum}")

even_sum = 0
odd_sum = 0
for i in range (101):
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i

print(f"The sum of all even is {even_sum}")
print(f"The sum of all odds is {odd_sum}")
 
```

```python
from countries import countries
land_countries =[country for country in countries if 'land' in country]
print("Countries containing land :",land_countries)


fruits =  ['banana', 'orange', 'mango', 'lemon']
reversed_fruits = []
for fruit in reversed(fruits):
    reversed_fruits.append(fruit)
print(f"Reversed fruits: {reversed_fruits}")



#using the countries_data file

with open("C:/Users/Habiba Isah/Desktop/30DaysOfPython/day_8/countries-data.py","r",encoding="utf-8") as file:
    data =eval(file.read())

if isinstance(data, list):
    languages = set(lang for country in data for lang in country['languages'])
    print("Total number of languages:", len(languages))
else:
    print("The data is not in the expected format (list of dictionaries).")
    

for language in sorted(languages):
    print(language)

from collections import Counter
language_counts = Counter(lang for country in data for lang in country['languages'])
most_spoken_languages = language_counts.most_common(10)
print("Ten most spoken languages:", most_spoken_languages)

#most populated countries
for country in data[:5]:  
    print(country)

populations = [(country['name'], country['population']) for country in data if 'population' in country]

most_populated_countries = populations[:10]
print("\nTen most Populated Countries:")
for country,population in most_populated_countries:
    print(f"{country}: {population}")
```
