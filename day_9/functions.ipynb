```python
import math

def add_two_numbers (a,b):
    return a + b
print(add_two_numbers(2,5))


def area_of_circle ():
    r = float(input("Enter the radius of the circle: "))
    area =  math.pi * r ** 2
    return area
print("Area of a cirle with radius 3:",area_of_circle())


def add_all_nums(*args):
   total = 0
   for arg in args:
       if not isinstance(arg, (int, float)):
           return"All arguments must be integers or floats!"
       total += arg
   return total
print("Sum fo 1-5:",add_all_nums(1,2,3,4,5))
print("Testing with invalid input :",add_all_nums(1,2,'a',4,5) )

def convert_celcius_to_farenheit(celcius):
    farenheit = (celcius * 9/5) + 32
    return farenheit
print("25°C in Fahrenheit is:",convert_celcius_to_farenheit(25))

def check_season(month):
    if month in ['September','October','November']:
        return 'Autumn'
    elif month in ['December','January','February']:
        return 'Winter'
    elif month in ['March','April','May']:
        return 'Spring'
    elif month in ['June','July','August']:
        return 'Summer'
    else:
        return 'Invalid month'
print(check_season ('September'))
print("The season for june:",check_season ('June'))
print(check_season ('Maris'))

def calculate_slope(x1,y1,x2,y2):
    if x1 == x2:
        return "The line is vertical and the slope is undefined"
    else:
        slope = (y2 - y1) / (x2 - x1)
        return slope
print("Slope of (1,2) and (3,4) is:",calculate_slope(1,2,3,4))

def solve_quadratic_eqn(a,b,c):
    discriminant = b ** 2 - 4 * a * c
    if discriminant < 0:
        return"No real roots"
    elif discriminant == 0:
        x = -b / (2 * a)
        return [x]
    else:
        x1 = (-b + math.sqrt(discriminant)) / (2 * a)
        x2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return [x1,x2]
print("Roots of 4x^2 + 4x + 1 are:",solve_quadratic_eqn(2,4,-1))
print("Testing with no real solutions :",solve_quadratic_eqn(1,2,3))


def print_list(list):
    for element in list:
        print(element)
    
print("List Elements:") 
print_list([1,2,3,4,5])

def reverse_list(arry):
    reversed_arry=[]
    for i in range(len(arry) - 1, -1, -1):
        reversed_arry.append(arry[i])
    return reversed_arry

print("Reversed list of [1,2,3,4,5]:",reverse_list([1,2,3,4,5]))
print("Reversed list of ['A','B','C']",reverse_list(['A','B','C',]))


def capitalize_list_items():
    list = ['apple','banana','cherry']
    return [item.upper() for item in list]
print("Capitalized list items:",capitalize_list_items())

def add_item(lst,item):
    lst.append(item)
    return lst
food_stuff = ['potatoe','Tomatoe','Mango','Milk']
print("After adding Meat:",add_item(food_stuff,'Meat'))
numbers = [2,3,7,9]
print("Adding 5 to list:",add_item(numbers,5))

def remove_item(lst,item):
    lst.remove(item)
    return lst
food_stuff = ['potatoe','Tomatoe','Mango','Milk']
print("After removing Mango:",remove_item(food_stuff,'Mango'))
numbers = [2,3,7,9] 
print("Removing 3 from list:",remove_item(numbers,3))  

def sum_of_numbers(n):
    return sum(range(n+1))
print("Sum of numbers up to 5:",sum_of_numbers(5))
print("Sum of numbers up to 10:",sum_of_numbers(10))    
print("Sum of numbers up to 100:",sum_of_numbers(100))

def sum_of_odds(n):
    return sum(i for i in range(n + 1) if i % 2 != 0)
print("Sum of odd numbers up to 5:",sum_of_odds(5))
print("Sum of odd numbers up to 10:",sum_of_odds(10))

def sum_of_evens(n):
    return sum(i for i in range(n + 1) if i % 2 == 0)
print("Sum of even numbers up to 5:",sum_of_evens(5))
print("Sum of even numbers up to 10:",sum_of_evens(10))
```

```python
import math
def evens_and_odds(n):
    evens = len([i for i in range (n+1) if i % 2 ==0])
    odds = len([i for i in range (n +1) if i % 2 !=0])
    return f"The numer of evens are {evens}. \nThe number of odds are {odds}"  
print(evens_and_odds(100)) 

def factorial(num):
    if num < 0:
        return "factorial does not exist for negative numbers."
    result = 1
    for i in range(1, num +1):
        result *= i
    return result
print("Factorial of 9:",factorial(9))

def is_empty(para):
    if len(para) == 0:
        return True
    else:
        return False
print("Checking if the parameter is empty?",is_empty(''))

def calculate_mean(lst):
    if not lst:
        return"List is empty"
    else:
        return sum(lst)/len(lst)

def calculate_median(lst):
    if not lst:
        return 'List is empty'
    lst.sort()
    n= len(lst)
    mid = n // 2
    if n % 2 ==0:
        return(lst[mid - 1] +lst[mid])/2
    return lst[mid]

from collections import Counter

def calculate_mode(lst):
    if not lst:
        return 'List is empty'
    freq = Counter(lst)
    max_freq = max(freq.values())
    mode = [key for key, value in freq.items() if value == max_freq]
    if len(mode) == len(lst):
        return 'No mode in the list'
    return mode

def calculate_range(lst):
    if not lst:
        return 'List is empty'
    return max(lst) - min(lst)

def calcuate_variance(lst):
    if not lst:
        return 'List is empty'
    mean = calculate_mean(lst)
    variance = sum((x - mean) ** 2 for x in lst) / len(lst)
    return variance    

def calculate_standard_deviation(lst):
    if not lst:
        return 'List is empty'
    variance = calcuate_variance(lst)
    return math.sqrt(variance)






numbers =list(range(1,51))

print("Mean :",calculate_mean(numbers))
print("Mean of an empty list:",calculate_mean([]))
print("Median:",calculate_median(numbers))
print("Mode:",calculate_mode(numbers))
print("Range:",calculate_range(numbers))
print("Variance:",calcuate_variance(numbers))
print("Standard deviation:",calculate_standard_deviation(numbers))
```

```python
def is_prime(n):
    if n < 1:
        return False
    for i in range (2, int(n ** 0.5) + 1):
        if n % i ==0:
            return False
    return True
print("Is 7 a prime",is_prime(7))
print("Is 9 a prime",is_prime(9))

def are_unique(lst):
    return len(lst) == len(set(lst))
print("Are all elements unique?",are_unique([1,2,3,4,5]))
print("Are all elements unique?",are_unique([1,2,3,4,4]))


def same_data_type(lst):
    return all(isinstance(item, type(lst[0])) for item in lst)
print(f"Are all items of the list of same type?{same_data_type([1,2,3,4,5])}")
print(f"Are all items of the list of same type?{same_data_type([1,2,3,4,'5',"a"])}")


import keyword
def valid_variable(var):
    if not var.isidentifier() or keyword.iskeyword(var):
        return False
    return True
print("Is 'hello' a valid variable name?",valid_variable('hello'))
print(f"Is my_bibex a valid variable name?{valid_variable('my_bibex')}")
print(f"Is 'for' a valid variable name? {valid_variable('for')}")
print(f"Is '1variable' a valid variable name? {valid_variable('1variable')}")



from collections import Counter

try:
    with open("C:/Users/Habiba Isah/Desktop/30DaysOfPython/day_8/countries-data.py","r",encoding="utf-8") as file:
        countries =eval(file.read())

except FileNotFoundError:
    print("The file was not found. Please check the file path.")
    countries = []
except Exception as e:
    print(f"An error occurred while loading the file: {e}")
    countries = []

def most_spoken_languages(n=10):
    if not countries:
        return "No data available"
    language_counts = Counter(
        language
        for country in countries 
        for language in country.get('languages',[])
    )
    return language_counts.most_common(n)
print("Ten most spoken languages:",most_spoken_languages(10))

def most_populated_countries(n=10):
    if not countries:
        return "No data available"
    return sorted(countries,key=lambda x: x['population'],reverse=True)[:n]
print("Ten most populated countries:",most_populated_countries(10))


```
