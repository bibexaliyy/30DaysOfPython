## 1.Map: 

 The map() function is a built-in function that takes a function and iterable as parameters.

- Used to transform elements in an iterable by applying function to each element.
- Returns a new iterable(e.g.,list in python  with explicit conversion).

Example:
  
  ```python
  numbers = [1,2,3,4,5]
  squares = map (labda x: x**2,numbers)
  print(list(squares))
 ```

### Filter:

The filter() function calls the specified function which returns boolean for each item of the specified iterable (list). It filters the items that satisfy the filtering criteria.

- Filters elements of an iterable based on a condititon(returns only elements that evaluate to ```True```)
  
Example:

```python
numbers = [1, 2, 3, 4, 5]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4]
```
### Reduce:

The reduce() function is defined in the functools module.Like map and filter it takes two parameters, a function and an iterable. However, it does not return another iterable, instead it returns a single value.

- Aggregates elements in an iterable into a single result by applying a function cumulatively.

Example:
```python
from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)  
```
## 2. Higher order functions:

This is a function that accepts another function as an argument or returns a function as it's result.

Example:
```python 
def apply_twice(func, x):
    return func(func(x))

print(apply_twice(lambda x: x + 2, 5))
```
### Decorator:
A special higher-order function that waps another function to modify or extend it's behavior.

Example:

```python
def decorator(func):
  def wrapper():
    print("Before the function call")
    func()
    print("After the function call")
    return wrapper
@decorator
def say_hello():
  print("Hello bibex!")

  say_hello()
```

````python
from collections import Counter
try:
    with open("C:/Users/Habiba Isah/Desktop/30DaysOfPython/day_8/countries-data.py","r",encoding="utf-8") as file:
        countries =eval(file.read())

except FileNotFoundError:
    print("The file was not found. Please check the file path.")
    countries = []
except Exception as e:
    print(f"An error occurred while loading the file: {e}")
    countries = []

sorted_by_name = sorted(countries,key= lambda x:x['name'])
print([country['name'] for country in sorted_by_name])


sorted_by_capital = sorted(countries,key= lambda x:x['capital'])
print([country['capital'] for country in sorted_by_capital if 'capital' in country])

sorted_by_population = sorted(countries,key= lambda x:x['population'],reverse=True)
print([country['name'] for country in sorted_by_population])

#most spoken countries
language_counts = Counter(language for country in countries for language in country['languages'])
most_spoken_languages = language_counts.most_common(10)
print(most_spoken_languages)

most_populated_countries = sorted(countries,key = lambda x: x['population'],reverse=True)[:10]
print([{'name':country['name'],'population':country['population']} for country in most_populated_countries])
````
